## 7) Creating a Function to Make Predictions

* Create a function to wrap up everythin we did
* Splitting training and test set and generating the predictions and then combining the prediction together
* So we will create a function this function will take our list of predictors, the core weather data frame and the regression model.

```python
def create_predictions(predictors, core_weather, reg):
    train = core_weather.loc[:"2020-12-31"]
    test = core_weather.loc["2021-01-01":]
    reg.fit(train[predictors], train["target"])
    predictions = reg.predict(test[predictors])
    error = mean_absolute_error(test["target"], predictions)
    combined = pd.concat([test["target"], pd.Series(predictions, index=test)], axis=1)
    combined.columns = ["actual", "predictions"]
    return error, combined
```

```python
from sklearn.model_selection import train_test_split

def create_predictions(predictors, core_weather, reg, test_size=0.2, random_state=42):
    # Split the data into training and testing sets
    train, test = train_test_split(core_weather, test_size=test_size, random_state=random_state)
    
    # Train the model
    reg.fit(train[predictors], train["target"])
    
    # Make predictions
    predictions = reg.predict(test[predictors])
    
    # Calculate the error
    error = mean_absolute_error(test["target"], predictions)
    
    # Combine actual and predictions
    combined = pd.concat([test["target"], pd.Series(predictions, index=test.index)], axis=1)
    combined.columns = ["actual", "predictions"]
    
    return error, combined
```

```python
# Predicting the temperature for tomorrow
def predict_tomorrow(core_weather, predictors, reg):
    # Train the model on the entire dataset
    reg.fit(core_weather[predictors], core_weather["target"])
    
    # Get the most recent data to predict the next day's temperature
    latest_data = core_weather.iloc[-1][predictors].values.reshape(1, -1)
    
    # Predict the temperature for tomorrow
    tomorrow_prediction = reg.predict(latest_data)[0]
    
    return tomorrow_prediction
```

```python
tomorrow_temp = predict_tomorrow(core_weather, predictors, reg)
print(f"Predicted temperature for tomorrow: {tomorrow_temp}")
```
